## TF solution: should also work??
(but it also optimized source energy which isn't necessary & it is more complicated)

```python
def tf_fit_CPVs_to_Yi_inv(BC):
    n_CPV = regressor.input_shape[1]
    regressor.trainable=False
    input_layer = tf.keras.layers.Input((n_CPV,))
    CPV_layer = tf.keras.layers.Dense(n_CPV, kernel_constraint=tf.keras.constraints.NonNeg(), name='CPV_bias')(input_layer)

    outs = regressor(CPV_layer) #['static_source_prediction']
    #outs = tf.slice(outs, (1,), (outs.shape[1]-1,))
    trained_inputs = tf.keras.models.Model(inputs=input_layer, outputs=outs)

    inputs = np.zeros(shape=(1,n_CPV))
    opt = tf.keras.optimizers.Adam(learning_rate=1e-7)
    trained_inputs.compile(optimizer=opt, loss='mse')
    trained_inputs.fit(inputs, {'static_source_prediction': np.asarray(BC).reshape(1,-1)}, epochs=1000)
    return trained_inputs.weights

Yi_cols = [col for col in df.columns if col.startswith('Yi')]
BC1 = df.iloc[0][['souener']+Yi_cols] #.filter(like='Yi')
BC2 = df.iloc[-1][['souener']+Yi_cols] #.filter(like='Yi')

weights = tf_fit_CPVs_to_Yi_inv(BC1)[1]
print(weights)
```
